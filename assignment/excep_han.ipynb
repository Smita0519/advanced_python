Write a program that prompts the user for a number and divides it by 2. Handle the ZeroDivisionError exception and print a friendly message instead of crashing.

```python
num = input("enter a number: ")

try:
    result = float(num)/2
    print(result)
except ZeroDivisionError as e:
    print("Math error", e)
```

Write a program that asks for the user's age. Check if the input is an integer and raise a ValueError if not. Print different messages based on age categories (e.g., child, adult, senior).

```python
age = input("Enter your Age: ")
try:
    if not age.isdigit():
        raise ValueError("Invalid Input")
    age = int(age)
    if age<15:
        print("Your Category: Child")
    elif age<60:
        print("Your Category: Adult")
    else:
        print("Your Category: Senior")
except ValueError as e:
    print("Value error:", e)
```
