Question 6: List Sanitization and Metrics                  
Given:                    
marks = [45, -3, 88, "NA", 102, None, 67]                     
Write a function that:                   
Removes invalid entries                                   
Returns:                    
Cleaned list                           
Number of invalid entries                        
Percentage of valid marks                   
Return values using a tuple.                 

```python
def sanitize_marks(marks):
    tot_entries = len(marks)
    cleaned = []
    
    for mark in marks:
        if isinstance(mark, (int, float)) and 0 <= mark <= 100: #is instance checks if the specific mark in the tuple matches the instance(int or float)
            cleaned.append(mark)

    invalid_cnt = tot_entries - len(cleaned)
    
    valid_per = round((len(cleaned) / tot_entries) * 100, 2)

    return (cleaned, invalid_cnt, valid_per)

marks = [45, -3, 88, "NA", 102, None, 67]
result= sanitize_marks(marks)
cleaned_list, invalid_count, valid_percentage = result

print("\n--- Sanitization Metrics ---")
print(f"Cleaned List of Valid Marks: {cleaned_list}")
print(f"Total Invalid Entries Found: {invalid_count}")
print(f"Percentage of Valid Marks: {valid_percentage}%")
```

Question 8: Prime Number Function
Write is_prime(n) and display prime numbers between 1 and 50.

```python
def is_prime(n):
    if n <= 1:
        return False
        
    # Check for divisors from 2 up to the square root of n; loop goes up to the square root (n**0.5) + 1 
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
            
    # No divisors found, so it is prime
    return True

prime_numbers = []

for number in range(1, 51):
    if is_prime(number):
        prime_numbers.append(number)

print("Prime Numbers between 1 and 50:")
print(prime_numbers)
```

Question 9: Function Composition
Write functions to:
Calculate total marks
Calculate average
Classify result (Distinction / Pass / Fail)
Then write a function that combines them and returns:
(name, total, average, result)
Explain why modular design is beneficial.

```python
#total marks
def calculate_total(marks):
    return sum(marks)

#average
def calculate_average(total_marks, num_subjects):
    return round(total_marks / num_subjects, 2)

#classifying the result
def classify_result(average):
    if average >= 80:
        return "Distinction"
    elif average >= 40:
        return "Pass"
    else:
        return "Fail"
    
#Composition Function
def process_student(name, marks):
    num_subjects = len(marks)
    total = calculate_total(marks)
    average = calculate_average(total, num_subjects)
    result = classify_result(average)
    return (name, total, average, result)


student1 = process_student("Derry", [90, 85, 70])
student2 = process_student("Merry", [55, 30, 65])
student3 = process_student("Gerry", [25, 35, 30])

print("Studentno.: ('Name', total marks, average, 'remarks')")
print(f"Student 1: {student1}")
print(f"Student 2:   {student2}")
print(f"Student 3: {student3}")
```

Question 10: Defensive Input Parsing              
Write a function that accepts a raw input string in the format:                           
"Ram|101|22|78, 88, 91"                             
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"                        
Tasks:                              
Extract name, roll number, age, and marks                         
Convert them into appropriate data types                    
Validate age (≥16) and marks (0-100)                         
Return the result as a dictionary using roll number as the key                
The function must not crash for malformed input.                   

```python
def parse_student_data(raw_input_string):

# Parses and validates student data. Returns: A dictionary {roll_number: {data}} if successful, or None if validation fails.
# Structural Check: Split by '|'
    parts = raw_input_string.split('|')       # converts into list
    
    if len(parts) != 4:
        print(f"ERROR: Input string '{raw_input_string}' has incorrect number of parts (expected 4).")  #invalid input
        return None

    name = parts[0].strip()      #separately store the input
    raw_roll = parts[1].strip()
    raw_age = parts[2].strip()
    raw_marks = parts[3].strip()

    # Roll Number Validation
    # .isdigit() checks if the string contains only digits.
    if raw_roll.isdigit():
        roll_num = int(raw_roll)
    else:
        print(f"VALIDATION FAILED: Roll number '{raw_roll}' is not an integer.")
        return None

    # Age Validation (>= 16)
    if raw_age.isdigit():
        age = int(raw_age)
        if age < 16:
            print(f"VALIDATION FAILED: Age {age} is below 16.")
            return None
    else:
        print(f"VALIDATION FAILED: Age '{raw_age}' is not an integer.")
        return None
        
    # Marks Validation (0-100)
    marks = []
    mark_list = raw_marks.split(',')
    
    for raw_mark in mark_list:
        raw_mark = raw_mark.strip()
        
        # Check if the mark string contains only digits
        if not raw_mark.isdigit():
            print(f"VALIDATION FAILED: Mark '{raw_mark}' is not a valid integer.")
            return None
        
        mark = int(raw_mark)
        if 0 <= mark <= 100:
            marks.append(mark)
        else:
            print(f"VALIDATION FAILED: Mark {mark} is outside the valid range (0-100).")
            return None

    student_record = {  #assemble
        'name': name,
        'age': age,
        'marks': marks,
    }
    
    return {roll_num: student_record}

Student1 = "Derry|101|22|78, 88, 91"
result = parse_student_data(Student1)
print(f"\nInput: {Student1}\nResult: {result}")

# Invalid Roll
input = "Gerry|A05|20|90, 80, 70"
resultI = parse_student_data(input)
print(f"\nInput: {input}\nResult: {resultI}")
```

Question 11: Type Stability and Conversion                      
Given the list:                            
data = [10, "20", 30.5, "abc", None, True]                           
Write a function that:                      
Separates numeric and non-numeric values                     
Converts numeric values to float                      
Returns both lists                       
Explain in comments why True behaves as a numeric value.                      

```python
def separate_and_convert(data_list):
    numeric_values = []
    non_numeric_values = []
    
    for item in data_list:
        # True/False (0/1) are technically integers,isinstance(item, float, bool) catches them.
        if isinstance(item, (int, float, bool)) and item is not None:
            #True behaves as 1 and False as 0 because the 'bool' class is a subclass of 'int' in Python.
            numeric_values.append(float(item)) #type conversion
        else:
            non_numeric_values.append(item)
            
    return numeric_values, non_numeric_values

data = [10, "20", 30.5, "abc", None, True]
nums, others = separate_and_convert(data)

print(f"Numeric: {nums}")
print(f"Non-numeric: {others}")
```

Question 13: Set-Based Enrollment Analysis
Given:

python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}

Find:

Students enrolled in exactly two subjects, 
Students enrolled in only one subject, 
Students enrolled in all subjects

```python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}

#Students enrolled in all subjects (Intersection of all three)
all_three = python & sql & statistics   # & (Bitwise AND)-> the result is 1 only if both bits are 1; INTERSECTION

# Students enrolled in exactly two subjects; find intersections of pairs and subtract those in all three
p_and_s = (python & sql) - all_three
s_and_st = (sql & statistics) - all_three
p_and_st = (python & statistics) - all_three
exactly_two = p_and_s | s_and_st | p_and_st       # | (Bitwise OR) -> result is 1 if at least one of the bits is 1; UNION

# Students enrolled in only one subject; Combine everything and subtract those in 2 or 3 subjects
all_students = python | sql | statistics        
only_one = all_students - exactly_two - all_three

print(f"Enrolled in all three: {all_three}")
print(f"Enrolled in exactly two: {exactly_two}")
print(f"Enrolled in only one: {only_one}")
```

Question 14: Dictionary Aggregation Logic
Given:

records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
Write a function that:

Computes average per student, 
Identifies top and bottom performers,
Returns results using a single dictionary

```python
def analyze(records):
    averages = {}
    
    # student's average
    for name, scores in records.items():
        avg = sum(scores) / len(scores)
        averages[name] = round(avg, 2)
    
    # top and bottom performers(max and min values)
    top_name = max(averages, key=averages.get)  #take each name,look up its value in the 'averages' dictionary, and compare those values.
    bottom_name = min(averages, key=averages.get)
    
    #Return everything in one dictionary
    return {
        "all_averages": averages,
        "top": {top_name: averages[top_name]},
        "bottom": {bottom_name: averages[bottom_name]}
    }

# Data
records = {
    "Amit": [78, 82, 91], #key: value
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

print(analyze(records))
```

Question 15: Nested Data Structure Analysis                        
Given a nested dictionary representing a school's grade book:                                     

school_data = {                              
    "Class A": {                                    
        "Amit": [78, 82, 91],                                
        "Sita": [88, 79, 85]                           
    },                         
    "Class B": {                         
        "Hari": [45, 52, 60],                        
        "Ram": [92, 88, 95]                      
    }                                
}                          
Write a function that:                           
                                
Calculates the average marks for each student,                                 
Finds the top performer in each class,                                    
Calculates the overall school average,                                 
Returns a dictionary with:[                             
class_toppers: {class_name: (student_name, average)}                                 
school_average: float                               
student_averages: {student_name: average}]                                  
Explain in comments why nested dictionaries are useful for hierarchical data.                           

```python
def analyze_school_data(school_data):
    student_averages = {}
    class_toppers = {}
    total_score_sum = 0
    total_score_count = 0

    # Nested dictionaries are useful for hierarchical data because they allow logical grouping (e.g., grouping students by class) which makes data retrieval organized and reflects real-world relationships.

    for class_name, students in school_data.items():
        current_class_topper = None
        highest_avg = -1

        for student_name, grades in students.items():
            # individual student average
            avg = sum(grades) / len(grades)
            student_averages[student_name] = round(avg, 2)

            # totals for the school average
            total_score_sum += sum(grades)
            total_score_count += len(grades)

            if avg > highest_avg:
                highest_avg = avg
                current_class_topper = (student_name, round(avg, 2))
        
        class_toppers[class_name] = current_class_topper

    # Calculate overall school average
    school_avg = total_score_sum / total_score_count if total_score_count > 0 else 0

    return {
        "class_toppers": class_toppers,
        "school_average": round(school_avg, 2),
        "student_averages": student_averages
    }

school_data = {
    "Class A": {"Amit": [78, 82, 91], "Sita": [88, 79, 85]},
    "Class B": {"Hari": [45, 52, 60], "Ram": [92, 88, 95]}
}

results = analyze_school_data(school_data)
print(results)
```
